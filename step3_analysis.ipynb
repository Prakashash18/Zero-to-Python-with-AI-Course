## Part A — the data as a list
A *list* is a row of boxes (the idea from Step 1). A `for` loop walks through it.


```python
days  = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
steps = [6200, 8100, 6400, 13420, 7300, 12500, 11200]

for i in range(len(days)):
    print(days[i], "->", steps[i], "steps")
```

## Part B — the big numbers


```python
total   = sum(steps)
average = total / len(steps)
print("Total:", total)
print("Average:", round(average))
print("Best day:", max(steps))
print("Quietest day:", min(steps))
```

## Part C — goal check (the `if` from Step 1)


```python
GOAL = 10000
for i in range(len(days)):
    if steps[i] >= GOAL:
        print(days[i], "hit the goal!")
```

## Part D — a real chart 📊
This is where Colab beats a simple editor: it draws proper graphs with `matplotlib`.


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
bars = plt.bar(days, steps)
plt.axhline(10000, linestyle="--", label="10,000 goal")
plt.title("My week in steps")
plt.ylabel("steps")
plt.legend()
plt.show()
```

## Part E — a reusable machine (the `function` from Step 1)


```python
def summarise(step_list):
    t = sum(step_list)
    a = round(t / len(step_list))
    return f"You walked {t} steps, averaging {a} a day."

print(summarise(steps))
```
